<h1>Signal Processing Code Explanation</h1>

This section elaborates on the provided Python code that demonstrates basic signal processing by generating and plotting a sine wave signal.

```python
from __future__ import division
import matplotlib.pyplot as plt  # Import for plotting
import numpy as np  # Import for mathematical operations

# Create the figure and axis
fig, ax = plt.subplots()

# Define the signal frequency and sampling rate
f = 1.0  # Hz, signal frequency
fs = 5.0  # Hz, sampling rate (fs >= 2 * f)

# Check for Nyquist criterion (optional but recommended)
if fs < 2 * f:
    raise ValueError("Sampling rate fs must be at least twice the signal frequency f (Nyquist criterion).")

# Generate the sample interval, symmetric for convenience
t = np.arange(-1, 1 + 1/fs, 1/fs)  # Time vector for sampling

# Generate the signal
x = np.sin(2 * np.pi * f * t)  # Signal in time domain

# Plot the signal
ax.plot(t, x, 'o-')  # Use standard ASCII quotes

# Label the plot
ax.set_xlabel('Time', fontsize=18)  # Correct quotes for label
ax.set_ylabel('Amplitude', fontsize=18)

# Display the plot
plt.xlim(-1,1)
plt.show()
```

    <li><strong>Imports</strong>: The code imports <code>matplotlib.pyplot</code> for plotting and <code>numpy</code> for numerical operations.</li>
    <li><strong>Figure and Axis Creation</strong>: <code>plt.subplots()</code> is called to create a new figure and axis for plotting.</li>
    <li><strong>Signal Definition</strong>: 
        <ul>
            <li><code>f</code> is defined as the frequency of the sine wave (1 Hz).</li>
            <li><code>fs</code> is the sampling rate (5 Hz), which is greater than twice the signal frequency to satisfy the Nyquist criterion.</li>
        </ul>
    </li>
    <li><strong>Nyquist Criterion Check</strong>: The code checks whether the sampling rate <code>fs</code> is at least twice the frequency <code>f</code> to prevent aliasing. If not, it raises a ValueError.</li>
    <li><strong>Time Vector Generation</strong>: The time vector <code>t</code> is generated using <code>np.arange()</code>, creating values from -1 to 1 seconds incremented by the sampling period (<code>1/fs</code>).</li>
    <li><strong>Signal Generation</strong>: The sine wave signal <code>x</code> is calculated using the formula <code>x(t) = sin(2πft)</code>, representing a continuous sine wave in the time domain.</li>
    <li><strong>Plotting</strong>: The signal is plotted against time using <code>ax.plot()</code>. Markers (<code>'o-'</code>) indicate the sampled points.</li>
    <li><strong>Plot Labeling</strong>: The x-axis and y-axis are labeled with 'Time' and 'Amplitude', respectively.</li>

</ol>

<h2>Mathematical Explanation</h2>

<p>The sine wave is defined mathematically as:</p>

<pre><code>x(t) = A ⋅ sin(2πft + φ)</code></pre>

<p>where:</p>
<ul>
    <li>A is the amplitude (not specified, assumed to be 1 here).</li>
    <li>f is the frequency (1 Hz).</li>
    <li>t is the time variable.</li>
    <li>φ is the phase offset (not used here, assumed to be ).</li>
</ul>

<p>The Nyquist theorem states that to properly sample a signal and reconstruct it without aliasing, the sampling frequency fs must be at least twice the maximum frequency present in the signal:</p>

<pre><code>fs ≥ 2f</code></pre>

<p>In this code, <code>fs</code> is set to 5 Hz, which satisfies the condition for <code>f = 1 Hz</code>.</p>

<p>This code effectively demonstrates the generation and visualization of a sine wave signal under the constraints of the Nyquist criterion, illustrating fundamental concepts in digital signal processing.</p>

<h1>Piecewise Linear Approximation</h1>

<p>This section discusses a Python implementation of a piecewise linear approximation process for a set of sampled points, including the necessary code and explanations.</p>

<h2>Code Implementation</h2>

```python
# Sample time points and corresponding values (for example)

# Initialize lists for piecewise domains and approximations
interval = []  # piecewise domains
apprx = []     # line on domains

# Build up points *evenly* inside of intervals
tp = np.hstack([np.linspace(t[i], t[i + 1], 20, endpoint=False) for i in range(len(t) - 1)])

# Construct arguments for piecewise
for i in range(len(t) - 1):
    interval.append(np.logical_and(t[i] <= tp, tp < t[i + 1]))  # Check which tp values fall in the current interval
    apprx.append((x[i + 1] - x[i]) / (t[i + 1] - t[i]) * (tp[interval[-1]] - t[i]) + x[i])  # Linear interpolation

# Use np.piecewise to create the piecewise linear approximation
x_hat = np.piecewise(tp, interval, apprx)  # Piecewise linear approximation
# Plotting the results
plt.figure(figsize=(10, 6))

# Plot original points
plt.plot(t, x, 'ro', label='Original Points')  # Red dots for original points

# Plot piecewise linear approximation
plt.plot(tp, x_hat, 'b-', label='Piecewise Linear Approximation')  # Blue line for approximation

# Adding labels and title
plt.title('Piecewise Linear Approximation', fontsize=18)
plt.xlabel('Time', fontsize=16)
plt.ylabel('Amplitude', fontsize=16)
plt.axhline(0, color='black', lw=0.5, ls='--')  # Horizontal line at y=0
plt.axvline(0, color='black', lw=0.5, ls='--')  # Vertical line at x=0
plt.legend()
plt.grid()
plt.show()
```

<h2>Code Breakdown</h2>

<ol>
    <li><strong>Initialization</strong>: Lists <code>interval</code> and <code>apprx</code> are initialized to store the domains and piecewise approximations.</li>
    <li><strong>Building Sample Points</strong>: <code>tp</code> is constructed to create evenly spaced points within each interval defined by the original time points <code>t</code>.</li>
    <li><strong>Piecewise Construction</strong>: A loop iterates over the intervals to:
        <ul>
            <li>Determine which points fall into the current interval using <code>np.logical_and()</code>.</li>
            <li>Calculate the linear interpolation for these points based on the adjacent values of the original signal.</li>
        </ul>
    </li>
    <li><strong>Creating the Piecewise Linear Approximation</strong>: <code>np.piecewise()</code> is used to construct the piecewise linear approximation using the previously determined intervals and their corresponding approximations.</li>
    <li><strong>Plotting the Results</strong>: A figure is set up, and both the original sampled points and the piecewise linear approximation are plotted.
        <ul>
            <li>Additional plot settings such as titles, labels, grid lines, and legends improve readability.</li>
        </ul>
    </li>
</ol>

<h2>Visualization</h2>

<p>The plot generated will display:</p>
<ul>
    <li><strong>Red dots</strong> representing the original sampled points.</li>
    <li>A <strong>blue line</strong> depicting the piecewise linear approximation, clearly showing how the approximation tracks the original signal between the sampled points.</li>
</ul>

<p>This code effectively demonstrates how to create and visualize a piecewise linear approximation for a given set of sampled data points, which is fundamental in various signal processing and graphical applications.</p>

<h1>Error Visualization with Piecewise Linear Interpolant</h1> <p>This section describes a Python code snippet that visualizes the errors associated with a piecewise linear interpolant compared to a sine wave.</p> <h2>Code Implementation</h2>

```python
# Create a new figure and axes
fig, ax1 = plt.subplots()

# Fill in the difference between the interpolant and the sine wave
ax1.fill_between(tp, x_hat, np.sin(2 * np.pi * f * tp), facecolor="red", alpha=0.5, label='Error Area')
ax1.set_xlabel('Time', fontsize=18)
ax1.set_ylabel('Amplitude', fontsize=18)

# Create a second y-axis sharing the same x-axis
ax2 = ax1.twinx()  # create clone of ax1

# Compute the squared error
sqe = (x_hat - np.sin(2 * np.pi * f * tp))**2  # Squared error between approximation and actual sine wave

# Plot the squared error
ax2.plot(tp, sqe, 'r', label='Squared Error')  # Plot squared error in red
ax2.axis(xmin=-1, xmax=1, ymax=sqe.max())  # Set x-axis limits and adjust y-axis to maximum squared error
ax2.set_ylabel('Squared error', color='r', fontsize=18)

# Set the title for the plot
ax1.set_title('Errors with Piecewise Linear Interpolant', fontsize=18)

# Show legends for both axes
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Display the plot
```

<h1>Mathematical Analysis of Sinc Interpolation</h1> <p>This section presents a Python code snippet that demonstrates sinc interpolation applied to a sampled signal. The process involves reconstructing a continuous signal from its discrete samples using the sinc function, which is crucial in signal processing.</p> <h2>Code Implementation</h2> 

```python
# Define the sampling rate
fs = 5.00  # Sampling rate in Hz

# Create the figure and axis
fig, ax = plt.subplots()

# Define time vectors
t = np.linspace(-1, 1, 100)  # Fine-grained time vector for smooth plot
ts = np.arange(-1, 1 + 1/fs, 1/fs)  # Sample points based on sampling rate

# Number of sinc coefficients based on the number of sampled points
num_coeffs = len(ts)

# Initialize the summation (reconstructed signal)
sm = 0  # Make sure it's a vector of zeros with the same shape as t

# Loop for sinc interpolation
for k in range(num_coeffs):
    sm += np.sin(2 * np.pi * (k / fs)) * np.sinc((t - ts[k]) * fs)

# Plot the reconstructed signal, original signal, and sampled points
ax.plot(t, sm, '--', label='Reconstructed Signal')  # Dashed line for sinc interpolated signal
ax.plot(t, np.sin(2 * np.pi * t), label='Original Signal')  # Solid line for original signal
ax.plot(ts, np.sin(2 * np.pi * ts), 'o', label='Sampled Points')  # Sample points as red dots

# Set the plot title and labels
ax.set_title('Sampling Rate = %3.2f Hz' % fs, fontsize=18)
ax.set_xlabel('Time', fontsize=16)
ax.set_ylabel('Amplitude', fontsize=16)

# Add legend
ax.legend()

# Show the plot
plt.show()
```

<h2>Analysis of the Code</h2> <ol> <li><strong>Define Sampling Rate</strong>: The variable <code>fs</code> represents the sampling rate in Hertz, determining how often samples are taken from the continuous signal.</li> <li><strong>Create Time Vectors</strong>: <ul> <li><code>t</code> is a fine-grained time vector that allows for a smooth plot of the continuous signal.</li> <li><code>ts</code> defines the actual sampled points based on the specified sampling rate.</li> </ul> </li> <li><strong>Initialize Summation</strong>: The variable <code>sm</code> is initialized to zero to accumulate the contributions from each sinc function during interpolation.</li> <li><strong>Sinc Interpolation Loop</strong>: A loop over the number of sample points is executed, where each iteration calculates the contribution of the corresponding sinc function to the reconstructed signal: <ul> <li>The term <code>np.sin(2 * np.pi * (k / fs))</code> evaluates the original signal at each sample point.</li> <li><code>np.sinc((t - ts[k]) * fs)</code> computes the sinc function that helps reconstruct the signal at each point in <code>t</code>.</li> </ul> </li> <li><strong>Plotting Results</strong>: The reconstructed signal (dashed line) is plotted against the original continuous signal (solid line) and sampled points (red dots), allowing for visual comparison.</li> <li><strong>Legend and Labels</strong>: The plot is annotated with titles and labels for both axes, and a legend is included to clarify the different elements represented.</li> </ol> <h2>Conclusion</h2> <p>This code snippet effectively illustrates the concept of sinc interpolation and its application in reconstructing a signal from discrete samples. The visual output provides insights into the accuracy of the sampling process and how well the piecewise reconstruction matches the original continuous signal.</p>

```python
# Define the sampling rate and frequency
fs = 5.00  # Sampling rate in Hz
f = 1.00   # Frequency of the sine wave

# Create the figure and axis
fig, ax1 = plt.subplots()

# Define time vector

# Number of sinc coefficients based on the number of sampled points
num_coeffs = len(ts)

# Initialize the summation (reconstructed signal)
sm = np.zeros_like(t)  # Initialize reconstructed signal

# Loop for sinc interpolation
for k in range(num_coeffs):
    sm += np.sin(2 * np.pi * (k / fs)) * np.sinc((t - ts[k]) * fs)

# Fill between original sine and reconstructed signal
ax1.fill_between(t, sm, np.sin(2 * np.pi * f * t), facecolor='red', alpha=0.5)

# Labels for amplitude
ax1.set_ylabel('Amplitude', fontsize=18)
ax1.set_xlabel('Time', fontsize=18)

# Create a second y-axis to plot squared error
ax2 = ax1.twinx()

# Compute squared error between original signal and reconstructed signal
sqe = (sm - np.sin(2 * np.pi * f * t))**2

# Plot squared error on the secondary axis
ax2.plot(t, sqe, 'r', label='Squared Error')

# Set axis limits for the squared error
ax2.set_ylim(0, sqe.max())

# Labels for squared error
ax2.set_ylabel('Squared Error', color='r', fontsize=18)

# Set plot title
ax1.set_title(r'Errors with Whittaker Interpolant', fontsize=18)

# Show the plot
plt.show()
```

<h1>Sine Reconstructed with Whittaker Interpolator</h1> <p>This section provides a Python code snippet that generates a plot illustrating the reconstruction of a sine wave using the Whittaker interpolator, based on sampled points.</p> <h2>Code Implementation</h2>

```python
# Define the time and sampling parameters
fs = 5.0  # Sampling rate (Hz)
t = np.linspace(-1, 1, 1000)  # Time vector for plotting from -1 to 1 with 1000 points

# Generate integer sampling coefficients based on the sampling rate
k = np.array(sorted(set((t * fs).astype(int))))  # Unique and sorted coefficients list

# Create the figure and axis
fig, ax = plt.subplots()

# Plot the individual Whittaker functions (sinc functions)
for i in range(len(k)):
    ax.plot(t, (np.sin(2 * np.pi * (k[i] / fs)) * np.sinc(k[i] - fs * t)), '--')

# Sum the Whittaker functions to generate the interpolant
whittaker_sum = np.sum(np.sin(2 * np.pi * (k[:, None] / fs)) * np.sinc(k[:, None] - fs * t), axis=0)
ax.plot(t, whittaker_sum, 'k-', label='Whittaker Interpolant')  # Whittaker interpolant in solid black

# Plot the sampled sine points
ax.plot(k / fs, np.sin(2 * np.pi * k / fs), 'ob', label='Sampled Points')  # Samples as blue circles

# Add labels and title
ax.set_xlabel('Time', fontsize=18)
ax.set_ylabel('Amplitude', fontsize=18)
ax.set_title('Sine Reconstructed with Whittaker Interpolator')

# Set axis limits to match the sine wave and sampling rate
ax.axis((-1.1, 1.1, -1.1, 1.1))

# Show the plot
plt.show()
```

 <h2>Explanation of the Code</h2> <ol> <li><strong>Define Time and Sampling Parameters</strong>: The sampling rate <code>fs</code> is set to 5 Hz, and a time vector <code>t</code> is created to cover the range from -1 to 1 with 100 points.</li> <li><strong>Generate Sampling Coefficients</strong>: Unique integer sampling coefficients are derived based on the time vector and sampling rate.</li> <li><strong>Create Figure and Axis</strong>: A figure and axis are created using <code>plt.subplots()</code> to prepare for plotting.</li> <li><strong>Plot Individual Whittaker Functions</strong>: A loop iterates through the sampling coefficients to plot the individual sinc functions corresponding to each coefficient.</li> <li><strong>Compute Whittaker Interpolant</strong>: The sum of the individual Whittaker functions is calculated to create the interpolated sine wave, which is then plotted on the same graph.</li> <li><strong>Plot Sampled Points</strong>: The sampled points of the sine wave are plotted as blue circles to indicate where the samples were taken.</li> <li><strong>Labels and Title</strong>: X and Y labels, as well as a title for the plot, are set to enhance readability.</li> <li><strong>Set Axis Limits</strong>: The plot's axis limits are specified to encompass the sine wave and its samples adequately.</li> <li><strong>Show the Plot</strong>: Finally, the plot is displayed using <code>plt.show()</code>.</li> </ol> <h2>Conclusion</h2> <p>This code effectively visualizes the reconstruction of a sine wave using the Whittaker interpolator based on sampled points. <h1>Computing Maximum Eigenvalues with a Kernel Function</h1> <p>This section presents a Python code snippet that defines a kernel function and computes the maximum eigenvalue of the kernel matrix for a given range of time-bandwidth products.</p> <h2>Code Implementation</h2>

```python
from numpy.linalg import eigvalsh

# Define the kernel function
def kernel(x, sigma=1):
    '''convenient function to compute kernel of eigenvalue problem'''
    x = np.asarray(x)  # Ensure x is treated as an array
    y = np.pi * np.where(x == 0, 1.0e-20, x)  # Avoid divide by zero and y = π * x
    return np.sin(sigma / 2 * y) / y  # Compute the kernel function & The {sigma / 2} term modifies the bandwidth of the kernel, controlling how tightly or loosely the kernel decays away from zero

# Define the number of steps for quantization
nstep = 100  # Quick and dirty integral quantization

# Define the time vector, t, over the range [-1, 1]
t = np.linspace(-1, 1, nstep)  # Time quantization

# Calculate the differential step size
dt = np.diff(t)[0]  # Differential step size

# Define a function to compute the maximum eigenvalue
def eigv(sigma):
    # Compute the kernel matrix and calculate its maximum eigenvalue
    return eigvalsh(kernel(t - t[:, None], sigma)).max()

# Define the range of sigma values (time-bandwidth product)
sigma = np.linspace(0.01, 4, 15)  # Range of sigma to consider

# Create the figure and axis for plotting
fig, ax = plt.subplots()

# Compute the maximum eigenvalues for each sigma and plot the result
ax.plot(sigma, dt * np.array([eigv(i) for i in sigma]), '-o')

# Add labels to the plot
ax.set_xlabel('Time-bandwidth product $\sigma$', fontsize=18)
ax.set_ylabel('Maximum Eigenvalue', fontsize=18)

# Set the y-axis limit
ax.axis(ymax=1.01)

# Add grid to the plot
ax.grid()

# Display the plot
plt.show()
```

<h2>Explanation of the Code</h2> <ol> <li><strong>Import Libraries</strong>: Import the necessary libraries: <code>numpy</code> for numerical operations, <code>matplotlib.pyplot</code> for plotting, and <code>eigvalsh</code> from <code>numpy.linalg</code> for eigenvalue computation.</li> <li><strong>Define the Kernel Function</strong>: The <code>kernel</code> function computes the kernel for the eigenvalue problem while avoiding division by zero.</li> <li><strong>Define Quantization Steps</strong>: The number of quantization steps is defined, set to 100 for this example.</li> <li><strong>Define Time Vector</strong>: A time vector <code>t</code> is created to cover the range from -1 to 1.</li> <li><strong>Compute Step Size</strong>: The differential step size <code>dt</code> is calculated from the time vector.</li> <li><strong>Compute Maximum Eigenvalue</strong>: The <code>eigv</code> function computes the maximum eigenvalue of the kernel matrix for a given <code>sigma</code>.</li> <li><strong>Define Range of Sigma Values</strong>: A range of <code>sigma</code> values is defined to explore the time-bandwidth product.</li> <li><strong>Create Plot</strong>: A plot is generated to visualize the relationship between the time-bandwidth product and the maximum eigenvalue.</li> <li><strong>Add Plot Labels</strong>: Labels for the x-axis and y-axis are added for clarity.</li> <li><strong>Display Plot</strong>: Finally, the plot is displayed using <code>plt.show()</code>.</li> </ol> <h2>Conclusion</h2> <p>This code effectively computes and visualizes the maximum eigenvalue of a kernel function defined by a sinc-like behavior. The plot illustrates how the maximum eigenvalue varies with different values of the time-bandwidth product, providing insights into the behavior of the kernel function in the context of eigenvalue problems.</p>

<h1>Computing and Plotting Eigenfunction for Kernel Matrix</h1> <p>This section provides a Python code snippet that computes the eigenvalues and eigenvectors of a kernel matrix and plots the eigenfunction corresponding to the maximum eigenvalue.</p> <h2>Code Implementation</h2>

```python
from numpy.linalg import eigh

# Define the time-bandwidth product sigma
sigma = 3  # Time-bandwidth product

# Compute the eigen-system of the kernel matrix
w, v = eigh(kernel(t - t[:, None], sigma))  # Compute eigenvalues `w` and eigenvectors `v`

# Identify the eigenfunction corresponding to the maximum eigenvalue
maxv = v[:, w.argmax()]  # Eigenfunction corresponding to the maximum eigenvalue

# Create a figure and axis for plotting
fig, ax = plt.subplots()

# Plot the eigenfunction corresponding to the maximum eigenvalue
ax.plot(t, maxv)

# Add labels and title to the plot
ax.set_xlabel('$t$', fontsize=22)  # x-axis label
ax.set_ylabel('$\psi_0(t)$', fontsize=22)  # y-axis label
ax.set_title('Eigenvalue=%3.4f; $\sigma$=%3.2f' % (w.max()*dt,sigma))  # Plot title

# Display the plot
plt.show()
```

<h1>Computing and Plotting Eigenfunction using a Kernel Function</h1> <p>This section contains a Python code snippet that defines a kernel function, computes the eigenvalues and eigenvectors of the kernel matrix, and plots the eigenfunction corresponding to the largest eigenvalue.</p> <h2>Code Implementation</h2>

```python
from numpy.linalg import eig

# Define the kernel function for the eigenvalue problem
def kernel_tau(x, W=1):
    '''Convenient function to compute kernel of eigenvalue problem.'''
    x = np.asarray(x)  # Ensure x is treated as an array
    y = np.pi * np.where(x == 0, 1.0e-20, x)  # Avoid divide by zero by replacing zeros with a small value
    return np.sin(2 * W * y) / y  # Compute sinc-like kernel

# Set parameters for time quantization
nstep = 300  # Number of steps for the time grid
t = np.linspace(-1, 1, nstep)  # Time grid for kernel function [-1, 1]
tt = np.linspace(-2, 2, nstep)  # Extended time grid [-2, 2]

# Define the time-bandwidth product and frequency parameter W
sigma = 5  # Time-bandwidth product
W = sigma / (2 * 2 * t.max())  # Define the frequency scaling based on sigma and time range

# Compute the eigenvalues and eigenvectors of the kernel matrix
w, v = eig(kernel_tau(t - tt[:, None], W))  # Compute the eigen-system for kernel_tau

# Identify the eigenvector corresponding to the largest eigenvalue
maxv = v[:, w.real.argmax()].real  # Extract the real part of the eigenvector for the maximum eigenvalue

# Plot the resulting eigenfunction
fig, ax = plt.subplots()
ax.plot(tt, maxv / np.sign(maxv[nstep // 2]))  # Normalize the eigenfunction for orientation
ax.set_xlabel(r'$t$', fontsize=24)  # x-axis label with LaTeX formatting
ax.set_ylabel(r'$\phi_{max}(t)$', fontsize=24)  # y-axis label with LaTeX formatting
ax.set_title(r'$\sigma=%d$' % (2 * W * 2 * t.max()), fontsize=26)  # Title with the time-bandwidth product

# Display the plot
plt.show()
```

<h2>Explanation of the Code</h2> <ol> <li><strong>Import Libraries</strong>: Import necessary libraries: <code>numpy</code> for numerical operations, <code>matplotlib.pyplot</code> for plotting, and <code>eig</code> from <code>numpy.linalg</code> for eigenvalue computations.</li> <li><strong>Define Kernel Function</strong>: The <code>kernel_tau</code> function computes a sinc-like kernel while avoiding division by zero.</li> <li><strong>Set Parameters</strong>: Define the number of quantization steps and create time grids <code>t</code> (for the kernel) and <code>tt</code> (for plotting).</li> <li><strong>Define Frequency Parameter</strong>: Calculate the frequency scaling factor <code>W</code> based on the time-bandwidth product <code>sigma</code>.</li> <li><strong>Compute Eigenvalues and Eigenvectors</strong>: The eigenvalues <code>w</code> and eigenvectors <code>v</code> of the kernel matrix are computed using the <code>eig</code> function.</li> <li><strong>Extract Maximum Eigenvalue's Eigenvector</strong>: The eigenvector corresponding to the largest real eigenvalue is identified.</li> <li><strong>Plot Eigenfunction</strong>: The eigenfunction is plotted, normalized for visualization purposes.</li> <li><strong>Add Labels and Title</strong>: x-axis and y-axis labels are defined along with a title indicating the value of the time-bandwidth product <code>sigma</code>.</li> <li><strong>Display Plot</strong>: Finally, the plot is displayed using <code>plt.show()</code>.</li> </ol> <h2>Conclusion</h2> <p>This code effectively computes and visualizes the eigenfunction corresponding to the largest eigenvalue of a kernel function. The plot illustrates the behavior of the eigenfunction over the specified time range, providing insights into the dynamics governed by the defined time-bandwidth product.</p>

<h1>Visualizing Eigenfunctions and DFT Representation</h1> <p>This section provides a Python code snippet that visualizes both the time-domain signal and its frequency-domain representation using a 3D diagram of the Discrete Fourier Transform (DFT).</p> <h2>Code Implementation</h2>

```python
from matplotlib import gridspec
from mpl_toolkits.mplot3d import Axes3D, art3d
from matplotlib.patches import FancyArrow
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

def kernel_tau(x, W=1):
    """Convenient function to compute kernel of eigenvalue problem."""
    x = np.asanyarray(x)
    y = np.pi * np.where(x == 0, 1.0e-20, x)  # avoid divide by zero
    return np.sin(2 * W * y) / y

def facet_filled(x, alpha=0.5, color='b'):
    """Construct 3D facet from adjacent points filled to zero."""
    a, b = x
    a0 = a * np.array([1, 1, 0])
    b0 = b * np.array([1, 1, 0])
    ve = np.vstack([a, a0, b0, b])  # create closed polygon facet
    poly = Poly3DCollection([ve])  # create facet
    poly.set_alpha(alpha)  # set transparency
    poly.set_color(color)
    return poly

def drawDFTView(X, ax=None, fig=None):
    """Draws 3D diagram given DFT matrix."""
    a = 2 * np.pi / len(X) * np.arange(len(X))
    d = np.vstack([np.cos(a), np.sin(a), np.abs(X).flatten()]).T
    if ax is None and fig is None:
        fig = plt.figure()
        fig.set_size_inches(6, 6)

    if ax is None:  # add ax to existing figure
        ax = fig.add_subplot(1, 1, 1, projection='3d')

    ax.axis([-1, 1, -1, 1])  # x-y limits
    ax.set_zlim([0, d[:, 2].max()])  # z-limit
    ax.set_aspect('auto')  # aspect ratio
    ax.view_init(azim=-30)  # camera view position
    a = FancyArrow(0, 0, 1, 0, width=0.02, length_includes_head=True)
    ax.add_patch(a)
    b = FancyArrow(0, 0, 0, 1, width=0.02, length_includes_head=True)
    ax.add_patch(b)
    art3d.patch_2d_to_3d(a)  # format 2D patch for 3D plot
    art3d.patch_2d_to_3d(b)
    ax.axis('off')

    sl = [slice(i, i + 2) for i in range(d.shape[0] - 2)]  # collect neighboring points
    for s in sl:
        poly = facet_filled(d[s, :])  # create polygon for each neighboring pair
        ax.add_collection3d(poly)

    # edge polygons
    ax.add_collection3d(facet_filled(d[[-1, 0], :]))  # connect last and first points
    ax.add_collection3d(facet_filled(d[[-2, -1], :]))  # connect second-to-last and last points

def drawInOut(X, v, return_axes=False):
    """Visualizes both the time-domain signal and its frequency-domain representation."""
    fig = plt.figure()
    fig.set_size_inches(8, 8)
    gs = gridspec.GridSpec(8, 6)

    ax1 = plt.subplot(gs[3:5, :2])  # 2D plot
    ax2 = plt.subplot(gs[:, 2:], projection='3d')  # 3D plot

    ax1.stem(np.arange(len(v)), v)  # Stem plot for input signal
    ymin, ymax = ax1.get_ylim()
    ax1.set_ylim(ymax=ymax * 1.2, ymin=ymin * 1.2)
    ax1.set_title('Signal')
    ax1.set_xlabel('n')
    ax1.tick_params(labelsize=8)

    drawDFTView(X, ax2)  # Draw 3D DFT view
    if return_axes:
        return ax1, ax2

# Example signal (sine wave)
Fs = 100  # Sampling frequency
t = np.arange(0, 1, 1/Fs)  # Time vector from 0 to 1 second
f = 5  # Frequency of sine wave
v = np.sin(2 * np.pi * f * t)  # Generate sine wave signal
X = np.fft.fft(v)  # Compute DFT of the signal

# Call the drawInOut function to visualize
drawInOut(X, v)
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()  # Display the plots
```

 <h2>Explanation of the Code</h2> <ol> <li><strong>Import Libraries</strong>: Import the necessary libraries, including <code>numpy</code>, <code>matplotlib</code>, and modules for 3D plotting.</li> <li><strong>Define Kernel Function</strong>: The <code>kernel_tau</code> function computes a kernel function to avoid zero division.</li> <li><strong>Define Facet Filling Function</strong>: The <code>facet_filled</code> function creates 3D facets from adjacent points filled to zero.</li> <li><strong>Draw DFT Function</strong>: The <code>drawDFTView</code> function visualizes the DFT of the signal in a 3D plot.</li> <li><strong>Draw In/Out Function</strong>: The <code>drawInOut</code> function creates both the time-domain and frequency-domain visualizations.</li> <li><strong>Generate Example Signal</strong>: An example sine wave signal is generated and its DFT is calculated.</li> <li><strong>Display Visualizations</strong>: Finally, the visualizations are displayed using <code>plt.show()</code>.</li> </ol> <h2>Conclusion</h2> <p>This code effectively combines time-domain and frequency-domain visualizations using both 2D and 3D plots, providing insights into the characteristics of the signal and its DFT. The DFT view offers a unique perspective on the frequency representation of the time-domain signal.</p>
